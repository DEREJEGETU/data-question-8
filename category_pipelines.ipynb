```python
# Function for preprocessing
def preprocess_paragraphs(paragraph_list):
    
    """Take a list of paragraphs and preprocess them for NLP.
    Results in nested lists of Bag-of-Words for each sentence"""
    
    sent_word_token = list()
    for p in paragraph_list:
        # Convert all words to lower case
        p_lower = p.lower()
        # Tokenize each sentence
        sent_tokens = sent_tokenize(p_lower)
        words = list()
        for sent in sent_tokens:
            # Tokenize each word in each sentence
            word_tokens = word_tokenize(sent)
            words.append(word_tokens)
        # Combine all the lists of tokenized objects
        sent_word_token.append(words)
        
    # This part can take some time to run, possibly refactor for greater speed or reorder so isn't looping so much
    for ind_1, p in enumerate(sent_word_token):
        for ind_2, sent in enumerate(p):
            # Remove stop words from each sentence
            sent_word_token[ind_1][ind_2] = [word for word in sent if word not in stopwords.words('english')]
            # Stem each word in each sentence
            sent_word_token[ind_1][ind_2] = [ps.stem(word) for word in sent]
            
    return(sent_word_token)
```

```python
full_sent_word_token = preprocess_paragraphs(paragraphs["paragraph_text"].unique())
```

```python
for para_ind, p in enumerate(full_sent_word_token):
    for sent in p:
        if any(word in sent for word in date_words):
            possible_dates = list()
            for ind, token in enumerate(sent):
                if token in stem_months:
                    if len(sent[ind+1]) == 4:
                        possible_dates.append(sent[ind:ind+2])
                    elif len(sent[ind+1]) <= 2:
                        possible_dates.append(sent[ind:ind+4])
                        
            print(para_ind, possible_dates)
            print("\n")
```
