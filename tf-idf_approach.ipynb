```python
categories = paragraphs.data_key_friendly_name.unique()

sub_df_list = list(range(len(categories)))
for ind, category in enumerate(categories):
    sub_df_list[ind] = paragraphs[paragraphs["data_key_friendly_name"] == category]
```

```python
para_text_lists = list()
for df in sub_df_list:
    para_text_lists.append(df["paragraph_text"].tolist())
```

```python
flat_docs = list()
for cat in para_text_lists:
    flat_docs.append(" ".join(cat))
```

```python
chars = ["[", ",", ".", ";", "@", "#", "?", "!", "&", "]", "+", "\\", "\`", "``", "[", "(", ")", "''"]
```

```python
for ind, p in enumerate(flat_docs):
    lower_case = p.lower()
    word_tokens = word_tokenize(lower_case)
    word_stop = [word for word in word_tokens if word not in stopwords.words("english")]
    word_chars = [word for word in word_stop if word not in chars]
    word_stem = [ps.stem(word) for word in word_chars]
    flat_docs[ind] = word_stem
```

```python
total_bow = [item for sublist in flat_docs for item in sublist]
```

```python
word_counts = Counter(total_bow)
```
